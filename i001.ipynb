```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset (replace with your dataset)
data = {
    'bedrooms': [3, 2, 4, 3, 2],
    'bathrooms': [2, 1, 3, 2, 1],
    'size_sqft': [2000, 1500, 2500, 1800, 1200],
    'price': [300000, 200000, 350000, 250000, 180000]
}

df = pd.DataFrame(data)

# Data exploration and visualization
# Pair plot to visualize pairwise relationships between features
sns.pairplot(df)
plt.show()

# Correlation heatmap
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Distribution of house prices
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], kde=True, bins=10, color='skyblue')
plt.xlabel('House Price ($)')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')
plt.show()

# Scatter plot of house prices vs. size_sqft
plt.figure(figsize=(8, 6))
sns.scatterplot(x='size_sqft', y='price', data=df, color='salmon', s=100)
plt.xlabel('Size (sqft)')
plt.ylabel('House Price ($)')
plt.title('House Price vs. Size')
plt.show()
```
